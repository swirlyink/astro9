```python
import numpy as np
%pylab inline
```

```python
#FIONA

E=300 #Electric field
B=1 #Magnetic field
m =1 #Mass
Ze =1 #Charge
t= np.linspace(0,1000)
v=E/B  #initial velocity in the x–y plane perpendicular to the magnetic field
theta = 2*np.pi #angle between v and the x-axis
w = Ze*B/m
a= m*E/Ze*(B**2)
x = a*w*t + ((v*np.sin(theta))/w)*(1-np.cos(w*t))-((a*v*np.cos(theta))/w)*np.sin(w*t)
y = ((v*np.sin(theta))/2)*np.sin(w*t)+ ((a -v*np.cos(theta))/w)*(1-cos(w*t))                                                                
```
